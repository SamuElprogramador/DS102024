## Ejercicio 1: Ortogonalidad

Definir una función en python que detecte si dos vectores son ortogonales

```python
import numpy as np
def son_ortogonales(v1,v2):
  #completá con tu código aquídef son_ortogonales(v1, v2):
   
    if len(v1) != len(v2):
        raise ValueError("Los vectores deben tener la misma dimensión.")
    
    producto_escalar = sum(a * b for a, b in zip(v1, v2))
    

    return producto_escalar == 0

vector1 = [1, 0, 0]
vector2 = [0, 1, 0]
print(son_ortogonales(vector1, vector2))
```

## Ejercicio 2: Desplazamiento

Definir una función en python que reciba como parámetro las cuatro esquinas de un cuadrilátero y las desplace 5 unidades hacia arriba y 7 unidades a la izquierda

```python
def esquina_cuadrilatero(esquinas):
   
    if len(esquinas) != 4:
        raise ValueError("Debe proporcionar exactamente cuatro esquinas.")
    
  
    esquinas_desplazadas = [(x - 7, y + 5) for x, y in esquinas]
    
    return esquinas_desplazadas

cuadrilatero = [(0, 0), (10, 0), (10, 10), (0, 10)]
cuadrilatero_desplazado = esquina_cuadrilatero(cuadrilatero)
print(cuadrilatero_desplazado) 
```
