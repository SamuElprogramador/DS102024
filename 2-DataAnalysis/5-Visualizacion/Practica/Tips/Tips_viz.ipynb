### Paso 6.  Crear una imagen con la relación de `total_bill`, `tip` y `size`.
#### Sugerencia: es sólo una función.

```python
import seaborn as sns
```

```python
def correlacion (tips, total_bill, tip, size):
  
  df_tips = tips[[total_bill, tip, size]]

  matriz_correlacion = df_tips.corr()

  plt.figure(figsize=(4, 3))

  sns.heatmap(matriz_correlacion, annot=True, cmap='coolwarm', fmt='.2f')

  plt.title('correlacion de bill tip y size')

  plt.show()


correlacion(tips, 'total_bill', 'tip', 'size')
```

### Paso 7. Presente la relación entre ``days`` y ``total_bill``

```python
day = tips['day']
total_bill = tips['total_bill']

plt.scatter(day, total_bill)
plt.xlabel('day')
plt.ylabel('total_bill')
plt.title('Correlacion entre day y total bill')
plt.show()
```

### Paso 8. Crear un gráfico de dispersión con el ``day`` como eje y y ``tips`` como eje x, diferenciar los puntos por ``sex``.

```python
columna_x = tips['tip']
columna_y = tips['day']
columna_sex = tips['sex']

for valor in columna_sex.unique():
    plt.scatter(columna_x[columna_sex == valor], columna_y[columna_sex == valor], label=valor)

plt.xlabel('tip')
plt.ylabel('day')
plt.title('correlación a 3')
plt.legend()  # Mostrar la leyenda
plt.show()
```
